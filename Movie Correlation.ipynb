```python
# Imports libaries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Reads in the data
df = pd.read_csv(r'C:\Users\Alex NZXT\Downloads\movies.csv')
```

```python
# Looks at the data

df.head()
```

```python
# Checks for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Deletes missing data

df = df.dropna()
```

```python
# Data types for columns

df.dtypes
```

```python
# Changes inconvenient data types into more convenient data types

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
# Creates a new column for the correct year of release

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
# Sorts gross column in descending order

df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# If needed, use below command to expand table
# pd.set_option('display.max_rows', None)
```

```python
# Drops any duplicates

df.drop_duplicates()
```

```python
# Are budget and gross correlated?
# Are companies and gross correlated?
```

```python
# Creates scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plots budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color": "green"})
```

```python
# Looks at correlation of df

df.corr(method = 'pearson') # pearson(default), kendall, spearman
```

```python
# High correlation between budget and gross
```

```python
# Creates correlation matrix with heatmap using df

correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Numerizes movie features of object type so that we can check correlation

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df
```

```python
# # Creates correlation matrix with heatmap using df_numerized

correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for All Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at corrrelation of df_numerized

df_numerized.corr()
```

```python
# Creates a list of correlations pairs

correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
# Sorts the list of correlation pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Returns the pairs with a correlation of over 0.5

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
